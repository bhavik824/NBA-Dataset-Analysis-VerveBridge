```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('C:/Users/Dell/OneDrive/Desktop/VervaBridge Tasks/achou-nba-draft-combine-measurements/nba_draft_combine_all_years.csv')

# Explore the data
print(df.head())
print(df.info())
print(df.describe())

# Analyze year-wise comparison

numeric_cols = df.select_dtypes(include='number').columns
df_yearwise = df.groupby('Year')[numeric_cols].mean()
print(df_yearwise.head())

# Visualize the data
plt.figure(figsize=(10, 6))
sns.heatmap(df_yearwise.corr(), annot=True)
plt.title('Correlation Matrix of Year-Wise Averages')
plt.show()

# Find key metrics and factors
# For example, we can analyze the relationship between height and wingspan
sns.scatterplot(x='Height (No Shoes)', y='Wingspan', data=df)
plt.title('Relationship between Height and Wingspan')
plt.show()

# Analyze the meaningful relationships between attributes
# For example, we can see if there is a correlation between standing reach and vertical leap
sns.scatterplot(x='Standing reach', y='Vertical (Max)', data=df)
plt.title('Relationship between Standing Reach and Vertical Leap')
plt.show()

```
